```python
#Dropping the Symbol Column because all of it is MARUTI
df=df.drop(['Symbol'],axis=1)

#Dropping all values where Series==BL
df=df[df['Series']!='BL']
df.reset_index(inplace=True)
```

```python
#1.2
pf=df['Close Price']
max_val=np.float64(0)
min_val=np.float64(100000)
avg=np.float64(0)
for i in range(len(df)-90,len(df)):
    if (pf[i] > max_val):
        max_val=pf[i]
    if (pf[i]<min_val):
        min_val=pf[i]
    avg=avg+pf[i]
```

```python
#For Maximum Value
print("The maximum closing price in the last 90 days is: Rs.{}".format(max_val))

#For Minimum Value
print("The minimum closing price in the last 90 days is: Rs.{}".format(min_val))

#For Maximum Value
print("The average closing price in the last 90 days is: Rs.{}".format(avg/90))
```

```python
def avgn(days):
    avgn=np.float64(0)
    for i in range(len(df)-days,len(df)):
        avgn=avgn+pf[i]
    return avgn/days
```

```python
def percChange(days):
    return (pf[len(df)-1]-pf[len(df)-days])/pf[len(df)-days]*100
    
```

```python
percChange_list=[7,14,30,90,180,365]
time_array =['1 Week','2 Weeks','1 Month','3 Months','6 Months','1 Year']
for time,days in zip(time_array,percChange_list):
    print('The Percentage Change for {} is: {}%'.format(time,percChange(days)))
```

```python
#1.6
df['Day_Perc_Change']=df['Close Price'].pct_change()*100
df.loc[0,'Day_Perc_Change']=0
```

```python
#1.7
df['Trend']='Null'
for i in range(len(df)):
    if((df['Day_Perc_Change'][i] > -0.5) & (df['Day_Perc_Change'][i] < 0.5) ):
        df.loc[i,'Trend']='Slight or No Change'
    elif((df['Day_Perc_Change'][i] > 0.5) & (df['Day_Perc_Change'][i] < 1 )):
        df.loc[i,'Trend']='Slight Positive'
    elif((df['Day_Perc_Change'][i] > -1) & (df['Day_Perc_Change'][i]< -0.5) ):
        df.loc[i,'Trend']='Slight Negative'
    elif((df['Day_Perc_Change'][i] > 1) & (df['Day_Perc_Change'][i]< 3) ):
        df.loc[i,'Trend']='Positive'
    elif((df['Day_Perc_Change'][i] > -3) & (df['Day_Perc_Change'][i]< -1) ):
        df.loc[i,'Trend']='Negative'
    elif((df['Day_Perc_Change'][i] > 3) & (df['Day_Perc_Change'][i]< 7) ):
        df.loc[i,'Trend']='Among Top Gainers'
    elif((df['Day_Perc_Change'][i] > -7) & (df['Day_Perc_Change'][i]< -3) ):
        df.loc[i,'Trend']='Among Top Losers'
    elif(df['Day_Perc_Change'][i] > 7):
        df.loc[i,'Trend']='Bull Run'
    elif(df['Day_Perc_Change'][i] < -7):
        df.loc[i,'Trend']='Bear Drop'
    
    
        
    
```

```python
#1.8
tren_df=df[['Total Traded Quantity','Trend']]
Trend_groups=tren_df.groupby('Trend')
```
